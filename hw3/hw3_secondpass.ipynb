### Import Everything

```python
import numpy as np
import matplotlib.pyplot as plt
import csv
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import MultinomialNB
from sklearn import metrics
```

### Load All Data

```python
# Raw Train File
with open('train.csv','r') as inputfile:
    traindata = np.array([row for row in csv.reader(inputfile,delimiter = ',',quotechar='"')])
traindata = traindata[1:,:]

# Utf-8 Train File
with open('train-utf8.csv','r') as inputfile:
    traindata_utf = np.array([row for row in csv.reader(inputfile,delimiter = ',',quotechar='"')])
traindata_utf = traindata_utf[1:,:]

# Raw Test File
with open('test.csv','r') as inputfile:
    testdata = np.array([row for row  in csv.reader(inputfile,delimiter = ',',quotechar='"')])
testdata = testdata[1:,:]

# utf-8 Test File
with open('test-utf8.csv','r') as inputfile:
    testdata_utf = np.array([row for row in csv.reader(inputfile,delimiter = ',', quotechar='"')])
testdata_utf = testdata_utf[1:,:]
```

### Define cross validation function for this assignment

```python
def LogisticCrossVal(data,nfolds=5,penalty='l1',C=1,maxngrams=1):
    # First Create folds list:
    fold = np.random.randint(0,nfolds,len(data))
    aucs = zeros(nfolds)
    vec = CountVectorizer(ngram_range=(0,maxngrams))
    for i in xrange(nfolds):
        X_train = vec.fit_transform(data[fold != i,2])
        Y_train = data[fold !=i, 0].astype('int')
        classifier = LogisticRegression(penalty=penalty,C=C).fit(X_train, Y_train)
        X_cv = vec.transform(data[fold ==i,2])
        Y_cv = data[fold ==i,0].astype('int')
        Y_cv_pred = classifier.predict_proba(X_cv)
        aucs[i] = metrics.auc_score(Y_cv,Y_cv_pred[:,1])
    return(aucs)
```

```python
print mean(LogisticCrossVal(traindata))
```

### Now let's wrap this in another function to try different values of regularization constant, try l1 and l2 regularization, and try the utf vs non-utf inputs

```python
def TestLogisticRegularization(data,n_lambdas=10,penalty='l1',maxngrams=1):
    # Set up constant array:
    C_values = linspace(0.001,1,n_lambdas)
    meanauc = zeros(n_lambdas)
    for i in xrange(n_lambdas):
        meanauc[i] = mean(LogisticCrossVal(data,3,penalty=penalty,C = C_values[i],maxngrams=maxngrams))
    return(zip(meanauc,C_values))
```

```python
res = TestLogisticRegularization(traindata,n_lambdas=20)
```

```python
plt.plot(array(res)[:,1],array(res)[:,0],linewidth=3)
plt.xlabel('C')
plt.ylabel('Mean AUC Score')
plt.title('Non-utf, L1 Reg, 3-fold CrossVal')
```

### I think it's easy enough to replace LogisticRegression with Naive Bayes

```python
def BayesCrossVal(data,nfolds=5,maxngrams=1):
    # First Create folds list:
    fold = np.random.randint(0,max(nfolds,10),len(data))
    aucs = zeros(nfolds)
    vec = CountVectorizer(ngram_range=(0,maxngrams),stop_words='english')
    for i in xrange(nfolds):
        X_train = vec.fit_transform(data[fold != i,2])
        Y_train = data[fold !=i, 0].astype('int')
        classifier = MultinomialNB().fit(X_train, Y_train)
        X_cv = vec.transform(data[fold ==i,2])
        Y_cv = data[fold ==i,0].astype('int')
        Y_cv_pred = classifier.predict_proba(X_cv)
        aucs[i] = metrics.auc_score(Y_cv,Y_cv_pred[:,1])
    return(aucs)
```

```python
print mean(BayesCrossVal(traindata,3,maxngrams=2))
```
